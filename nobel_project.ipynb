```python
# Kategorilere göre kazananları filtreleme 
def filter_winners_by_category(kategori):
    return grup_df3[grup_df3['kategori'] == kategori]

# Ödül kategorileri listesi
categories = grup_df3['kategori'].unique()

# Her bir ödül kategorisi için ayrı grafik oluşturma
for kategori in categories:
    winners = filter_winners_by_category(kategori)
    
    plt.figure(figsize=(12, 8))
    
    # Yas Dağılımı
    plt.subplot(2, 2, 1)
    sns.histplot(data=grup_df3, x='yas')
    plt.title('Yas Dağılımı')
    plt.xlabel('Yas')
    plt.ylabel('Kazanan Sayısı')
    
    ## Ülke Dağılımı
    plt.figure(figsize=(12, 8))
    sns.barplot(x='organizasyon_ülke', y='sayı', data=grup_df3, palette='viridis')
    plt.title('Ülke Dağılımı', fontsize=16)
    plt.xlabel('Ülke', fontsize=14)
    plt.ylabel('Frekans', fontsize=14)
    plt.xticks(rotation=65, ha='right')
    plt.grid(axis='y', linestyle='--', alpha=0.7)

    plt.suptitle(f'{kategori} Nobel Ödülü Kazananların Özellikleri', fontsize=16)
    plt.tight_layout()
    plt.show()
```
